## Ejercicio 1
Calcular la evolución de los primeros pasos (10 si son a mano y 100 si son con un programa) de los
siguiente mapas, partiendo de la condición inicial x 0 = 0.3:
1. $ x_{n+1} = \frac{4}{5} (1 − x_{n}) $
2. $ x_{n+1} = 1 − x^{2}_n $
3. $ x_{n+1} = 4x_{n} (1 − x_{n}) $

```python
def map1(x):
    return (1-x) * 4 / 5

def map2(x):
    return (1 - x ** 2)

def map3(x):
    return 4 * x * (1 - x)
```

```python
map1_first_100_steps = []
x0 = 0.3
x = x0
for idx in range(100):
    x = map1(x)
    map1_first_100_steps.append(x) 

plt.plot(map1_first_100_steps,'o-')
```

## Ejercicio 10
El módelo clásico de propagación de epidemias es el SIR, que divide la población en tres grupos:
suceptibles (S), infectados (I) y recuperados (R). Los suceptibles son los que todavia no se infectaron
y se pueden infectar, los infectados luego de un tiempo pasan a formar parte de la poblacion de
recuperados y los recuperados no se reinfectan. Como la suma S + I + R tiene que ser la de la
población total nuestro sistema dinámico va a quedar descripto por sólo dos variables (S e I). A
todo momento el futuro del sistema queda determinado por los valores de estas dos variables. Vamos
a trabajar con valores de población normalizada por la cantidad total de habitantes, es decir S e I
varían entre 0 y 1, y además R = 1 − S − I. Al principio toda la población es suceptible y hay una
fracción muy pequeña de infectados. En cada paso de tiempo parte de los suceptibles pasa a estar
infectada y esa cantidad es función del producto SI (pensar por qué yendo a los extremos de que no
haya infectados o de que todos estén infectados) y de un parametro de control β que incluye cosas
tales como cuán contagiosa es la enfermedad, cuán en contacto está la población infectada con la
suceptible. Por otra parte en cada paso de tiempo una fracción γ de los infectados se recupera y ese es
un parámetro fijo. Las ecuaciones SIR escritas como mapa (discreto en el tiempo) se pueden escribir
como:
S n+1 = S n − βS n I n (1)
I n+1 = I n + βS n I n − γI n (2)
Escribir un programa para calcular la evolución del sistema partiendo de la condición inicial S 0 =
0.99, I n = 0.01 (es decir un 1% de infectados) con γ = 0.1 y comparar tres valores del parámetro β:
0.2 (epidemia suave), 0.5 (epidemia grave), 1 (virulenta). Que forma tiene la curva de infectados a lo
largo del tiempo? Que significado tiene en este contexto la idea de ’achatar la curva’ en la pandemia
de COVID19?

```python
def iteration(susceptible, infected, beta, recovered_ratio):
    sus = susceptible - beta * susceptible * infected
    inf = infected + beta *  susceptible * infected - recovered_ratio * infected
    return (sus, inf)
```

```python
T = 100
beta = [0.2, 0.5, 1]
s0 = 0.99
initial_infected = 0.01
recovered_ratio = 0.1

for b in beta:
  sus =  s0
  infected = initial_infected
  betan = []
  for t in range(T):
    sus,infected = iteration(sus, infected, b, recovered_ratio)
    betan.append(infected)
  plt.plot(betan,'-o');
```

## Ejercicio 11
Podemos escribir un mapa de dos variables x n+1 = f 1 (x n , v n ), v n+1 = f 2 (x n , v n ) como aproximación
en los tiempos discretos t = n ∗ ∆t de la evolución de la posición x(t) y la velocidad v(t) del sistema de
la masa en el resorte, definiendo la velocidad como la diferencia de posición en dos pasos de tiempo
consecutivos (x n+1 − x n ) dividida por el intevalo de tiempo ∆t, es decir v n = (x n+1 − x n )/∆t. De
esta definición obtenemos la función f 1 despejando x n+1 . Para la función f 2 aplicamos la segunda ley
de Newton que nos dice que la variación de la velocidad en el intevalo de tiempo ∆t (aceleración) es
proporcional a la fuerza (la constante de proporcionalidad es la masa que acá tomamos igual a 1 para
simplificar). La fuerza está dada por la ley de Hooke como −K ∗ x n . En ese caso la aceleración sería:
(v n+1 − v n )/∆t y al igualarla con la fuerza obtenemos f 2 despejando v n+1 . El sistema quedaría escrito
entonces (verificar despejando las variables) de la siguiente forma*:

$ x_{n+1} = x_{n} + v_{n}∆t $



$ v_{n+1} = v_{n} − Kx_{n} ∆t $

Escribir un programa para calcular la evolución del sistema usando ∆t = 0.01 y K = 40, partiendo de
x 0 = 1 v 0 = 0. Usar al menos 100 iteraciones. Graficar la evolución de x y v a lo largo de tiempo y el
retrato de fases de en el espacio (x, v). Se cierran las trayectorias como en el sistema físico real? Por
qué? Probar ahora con un ∆t más chico. Tener en cuenta que cuanto menor sea ∆t (mayor frecuencia
de sampleo por usar una analogía con audio) hay que calcular una mayor cantidad de iteraciones para
’simular’ la misma cantidad de tiempo. Qué sucedería si ∆t se hace infinitamente pequeño?
* Esto equivale al esquema de Euler explícito para integrar un sistema de ecuaciones diferenciales
ordinarias

```python
def iteration_for_ex_11(position, velocity, delta_time = 0.01, constant_k= 40):
    pos = position + velocity * delta_time
    vel = velocity - constant_k * position * delta_time
    return (pos, vel)
```

```python
initial_position = 1
initial_velocity = 0
time = np.arange(0,2,0.01)
positions = np.zeros(time.shape)
positions[0] = initial_position
velocities = np.zeros(time.shape)
velocities[0] = initial_velocity
for x in range(1,len(time)):
    (positions[x],velocities[x]) = iteration_for_ex_11(positions[x-1],velocities[x-1])

plt.plot(positions,velocities)
```
